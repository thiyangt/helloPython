```python
import numpy as np
import pandas as pd
from tsfeatures import tsfeatures

periods = 11
ind = pd.date_range(start='2000-01-01', periods=periods, freq='D')
vals = np.random.rand(periods)

df = pd.DataFrame({'ds':ind, 'y':vals, 'unique_id':1})
print(f'Inferred freq: {pd.infer_freq(df.ds)}')
print(df.head())
```
