## filter

```python
# filter(gapminder, lifeExp < 50) 
gapminder[gapminder['lifeExp'] < 50]
```

```python
#filter(gapminder, country == "Sri Lanka")
gapminder[gapminder['country']=="Sri Lanka"]
```

## select

```python
#select(gapminder, year:gdpPercap)
gapminder[['year','lifeExp','pop', 'gdpPercap']]
```

```python
 gapminder.iloc[:, 1:3]
```

```python
gapminder.drop('year', 1)
```

```python
gapminder.drop('year', axis=1)
```

```python
gapminder.drop(['year', 'country'], axis=1)
```

## mutate

```python
# gapminder %>% mutate(gdp = pop * gdpPercap)
gapminder.assign(gdp = lambda x: (x['pop'] * x['gdpPercap']))
```

```python
gapminder.assign(pop100 = lambda x: (x['pop'] * 100))
```

## summarize

```python
gapminder.mean()
```

```python
gapminder.describe()
```

```python
gapminder.agg(['mean', 'count'])
```

```python
gapminder.describe(include=[np.number])
```

```python
gapminder.describe(include='all') 
```

```python
gapminder.describe(include=[object]) 
```

```python
gapminder.describe(exclude=[np.number]) 
```

## groupby

```python
gapminder.groupby(['continent']).mean()
```

```python
gapminder.groupby(['continent']).agg(['mean', 'count'])
```

```python
gapminder.groupby(['continent']).describe()
```

```python
gapminder.groupby(['continent'])['pop'].mean()
```

```python
df2=gapminder.groupby(['continent'])['pop'].describe()[['count', 'mean']]
df2
```
