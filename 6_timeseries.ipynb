```python
#airpassenger["month"] = airpassenger['Month'].map(lambda x: x.month)
#airpassenger["year"] = airpassenger['Month'].map(lambda x: x.year)
airpassenger['year'] = airpassenger['Month'].dt.year
airpassenger['month'] = airpassenger['Month'].dt.month
```

## Visualizing birth dataset

```python
birth_reduced = pd.read_csv('birth_reduced.csv')
```

```python
birth_reduced.info()
```

```python
birth_reduced.head()
```

```python
# count number of observations for each disease by year of birth
yearly_counts = birth_reduced.groupby(['year_of_birth', 'disease'])['disease'].count()
yearly_counts.head()
```

```python

yearly_counts.info()
```

```python
yearly_counts = yearly_counts.reset_index(name='counts')
yearly_counts.head()
```

```python
yearly_counts.info()
```

```python
# plot all counts by year
ggplot(yearly_counts, aes(x='year_of_birth', y='counts'))+ geom_line()
    
```
